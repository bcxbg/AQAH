```python
import json

import re
```

```python
def get_words(sentence):
    cleaned_sentence = re.sub(r'[^\w\s]', '', sentence)  # 使用正则表达式移除标点符号
    words = cleaned_sentence.split()
    #words = sentence.split()
    return words,len(words)
```

```python
dataset_path = "/data/dtt/project/LLaVA-main/hallucination/code/original_details/6_captions.json"

with open(file = dataset_path,encoding = 'utf-8',mode = 'r') as file:
    temp_dict = json.load(file)

top_k_captions = []
beam_captions = []

for image_name,image_files in temp_dict.items():
    if image_name[-4:] != ".jpg":
        continue
    if image_files['top_k_text'] != None:
        top_k_captions.extend(image_files['top_k_text'])
    if image_files['beam_text:'] != None:
        beam_captions.extend(image_files['beam_text:'])
    
top_k_count = len(top_k_captions)
beam_count = len(beam_captions)

total_captions = []
total_captions.extend(top_k_captions)
total_captions.extend(beam_captions)
total_count = top_k_count+beam_count

total_number=  0
for i in range(len(total_captions)):
    temp_sentence = total_captions[i]
    if temp_sentence == None:
        total_count-=1
        continue
    words,words_number = get_words(temp_sentence)
    print(words)
    total_number += words_number
print(total_number)
print(total_count)
print(round(total_number/total_count,2))
```
